#Machine Learning - Breast Cancer Wisconsin (Diagnostic)

```
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

breast_diagnostic = pd.read_csv("drive/My Drive/Colab Datasets/data.csv")
breast_diagnostic.head()
```

#Roteiro Aprendizado Supervisionado

1. Passo I:   Escolha uma classe de modelo importando do Scikit.
2. Passo II:  Escolha os hiperparâmetros do modelo instaciando a classe escolhida no Passo I.
3. Passo III: Organize os dados em uma *feature matrix* e um *target array*.
4. Passo IV:  Ajuste o modelo invocando o método **fit()**.
5. Passo V:   Aplique o modelo aos novos dados usando **predict()**.

##Por Decision Tree:

```
from sklearn.tree import DecisionTreeClassifier   # Passo I: escolha do modelo
model_decisiont = DecisionTreeClassifier()        # Passo II: escolha dos hiperparâmetros e instanciação
```

```
#Passo III_I: Feature Matrix
X_breast = breast_diagnostic.loc[:, [
                                      'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
                                      'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se', 'perimeter_se', 'area_se',
                                      'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst',
                                      'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst',
                                      'symmetry_worst', 'fractal_dimension_worst'
                                     ]]
X_breast.head()
```

```
#Passo III_II: Target Array
y_breast = breast_diagnostic.diagnosis
y_breast.head()
```

```
#Passo III_III: Separação do Banco de Dados
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_breast, y_breast, random_state=2020)
```

```
#Passo IV: Ajuste o modelo aos dados
model_decisiont.fit(X_train, y_train)
```

```
#Passo V: Faça predições em dados novos
y_model_decisiont = model_decisiont.predict(X_test)
y_model_decisiont
```

```
y_test.to_numpy()
```

```
#Verificando a acurácia do modelo escolhido
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model_decisiont)
```

##Por Naive Bayes:

```
from sklearn.naive_bayes import GaussianNB    # Passo I: escolha do modelo
model_bayes = GaussianNB()                    # Passo II: escolha dos hiperparâmetros e instanciação
```

```
#Passo IV: Ajuste o modelo aos dados
model_bayes.fit(X_train, y_train)
```

```
#Passo V: Faça predições em dados novos
y_model_bayes = model_bayes.predict(X_test)
y_model_bayes
```

```
y_test.to_numpy()
```

```
#Verificando a acurácia do modelo escolhido
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model_bayes)
```

##Por Redes Neurais (Neural Network):

```
from sklearn.neural_network import MLPClassifier                       # Passo I: escolha do modelo
model_neuraln = MLPClassifier(solver='adam', activation='relu')        # Passo II: escolha dos hiperparâmetros e instanciação
```

```
#Passo IV: Ajuste o modelo aos dados
model_neuraln.fit(X_train, y_train)
```

```
#Passo V: Faça predições em dados novos
y_model_neuraln = model_neuraln.predict(X_test)
y_model_neuraln
```

```
y_test.to_numpy()
```

```
#Verificando a acurácia do modelo escolhido
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_model_neuraln)
```

##Cross Validation:

```
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

model_neighbors = KNeighborsClassifier(n_neighbors=1)
cross_val_score(model_neighbors, X_breast, y_breast, cv=5)
```
