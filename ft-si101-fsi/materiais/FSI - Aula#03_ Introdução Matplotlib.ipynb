#Matplotlib - Gráficos

```
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

wine_reviews = pd.read_csv ("winemag-data-130k-v2.csv", index_col=0)
wine_reviews.head(7)
```

* value_counts -> organiza em ordem de frequência dos elementos (por categoria)

```
wine_reviews.province.value_counts().head(10).plot.bar()
```

* len (wine_reviews) -> exibe o total de valores de linhas

```
(wine_reviews.province.value_counts().head(10) / len(wine_reviews)).plot.bar()
```

* Sort_index() -> organiza os elementos em ordem

* Gráfico de barras -> ilegível quando o número de categorias aumenta

```
wine_reviews.points.value_counts().sort_index().plot.bar()
```

```
wine_reviews.points.value_counts().sort_index().plot.line()
```

```
wine_reviews.points.value_counts().sort_index().plot.area()
```

```
wine_reviews.points.plot.hist()
```

- sort_values -> ordem decrescente

```
wine_reviews.points.value_counts().sort_values().plot.bar()
```

* Histogramas -> exibem mais elementos por categoria

```
wine_reviews.price.plot.hist()
```

```
wine_reviews[wine_reviews.price < 200].price.plot.hist()
```

* Pizza

```
wine_reviews.province.value_counts().head(10).plot.pie()
```

* Gráficos com duas variáveis

```
wine_reviews[wine_reviews.price < 100].plot.scatter(x='price', y='points')
```

```
wine_reviews[wine_reviews.price < 100].sample(100).plot.scatter(x='price', y='points')
```

```
wine_reviews[wine_reviews.price < 100].plot.hexbin(x='price', y='points', gridsize=20)
```
