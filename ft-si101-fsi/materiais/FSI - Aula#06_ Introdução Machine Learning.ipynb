#Machine Learning - Regressão Linear

```
import matplotlib.pyplot as plt
import numpy as np #Análise numérica
```

```
#f(x) = a * x + b
#x é a feature (o que temos dentro do banco de dados)
rng = np.random.RandomState(2020)
x = 10 * rng.rand(50)
x
```

```
#y é o rótulo/target/label (o que queremos descobrir)
y = 2 * x + rng.randn(50)
y
```

```
plt.scatter(x, y)
```

```
from sklearn.linear_model import LinearRegression #Passo I: importar o modelo Regressão Linear
model = LinearRegression(fit_intercept=True)      #Passo II: configurar os hyperparâmetros
                                                  #Passo III: organizar os dados
                                                  #Passo IV: Ajustar o modelo aos dados (modelo treinado)
                                                  #Passo V: fazer predições em dados novos
```

```
#Passo III: organizar os dados
X = x[:, np.newaxis]
X
```

```
#Passo IV: Ajustar o modelo aos dados (modelo treinado)
model.fit(X, y)
```

```
model.coef_
```

```
model.intercept_
```

```
# f(x) = a * x + b
# a -> coef
# b -> intercept
result = np.dot(X, model.coef_) + model.intercept_
result
```

```
# predict = f(x)= a * x + b = result
model.predict(X)
```

```
plt.scatter(X, y, label='Valores Originais')
plt.plot(X, result, 'r--', label='Valores Obtidos')
```

```
#Passo V: fazer predições em dados novos
xfit = np.linspace(-1, 11)
xfit = xfit[:, np.newaxis]
yfit = model.predict(xfit)

plt.scatter(x, y)
plt.scatter(xfit, yfit)
```
