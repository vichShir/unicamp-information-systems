```
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
df_train = pd.read_csv("drive/My Drive/Colab Datasets/train.csv")
df_test = pd.read_csv("drive/My Drive/Colab Datasets/test.csv")
df_test
```

```
df_train.info()
```

#Filtro por sobreviventes

```
sns.countplot(x='Survived', data = df_train)
```

```
from google.colab import files

df_test["Survived"] = 0
df_test.loc[:, ['PassengerId', 'Survived']].to_csv('no_survived.csv', index = False)
files.download('no_survived.csv')
```

```
sns.countplot(x='Embarked', data = df_train)
```

#Filtro por Sexo == female

```
sns.factorplot(x='Survived', col='Sex', kind='count', data = df_train)
```

```
df_train.groupby(['Sex']).Survived.sum()
```

```
df_test['Survived'] = df_test.Sex == 'female'
df_test['Survived'] = df_test.Survived.apply(lambda x: int(x))

df_test.loc[:, ['PassengerId', 'Survived']].to_csv('woman_survived.csv', index = False)
files.download('woman_survived.csv')
```

#Filtro por Pclass == 1

```
sns.factorplot(x='Survived', col='Pclass', kind='count', data = df_train)
```

```
df_test['Survived'] = df_test.Pclass == 1
df_test['Survived'] = df_test.Survived.apply(lambda x: int(x))

df_test.loc[:, ['PassengerId', 'Survived']].to_csv('class_survived.csv', index = False)
files.download('class_survived.csv')
```

#Filtro por Embarked == C

```
sns.factorplot(x='Survived', col='Embarked', kind='count', data = df_train)
```

```
df_test['Survived'] = df_test.Embarked == "C"
df_test['Survived'] = df_test.Survived.apply(lambda x: int(x))

df_test.loc[:, ['PassengerId', 'Survived']].to_csv('french_survived.csv', index = False)
files.download('french_survived.csv')
```
