# Usando o wine_reviews

```
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

wine_reviews = pd.read_csv("drive/My Drive/Colab Datasets/winemag-data-130k-v2.csv", index_col=0)
wine_reviews.head()
```

# Função map()

```
#-> média igual/próximo a zero (wine_reviews.points.map(lambda p : p - mean)
mean = wine_reviews.points.mean()

wine_reviews.points.map(lambda p : p - mean)
```

```
wine_reviews.points.mean()
```

# Função apply()

```
mean = wine_reviews.points.mean()

def remean_points(line):
  line.points = line.points - mean
  return line

wine_reviews.apply(remean_points, axis='columns')
```

# Exemplificando funcionamento do apply()

```
workbook = pd.DataFrame (
    {
        'Yes' : [50, 21, 11],
        'No'  : [131, 2, 10],
        'Maybe' : [45, 22, 1]
    }, index=['Produto A', 'Produto B', 'Produto C']
)

workbook
```

```
workbook.apply(lambda x : x[1] + 10, axis='rows')
```

```
workbook.apply(lambda x : x[1] + 10, axis='columns')
```
